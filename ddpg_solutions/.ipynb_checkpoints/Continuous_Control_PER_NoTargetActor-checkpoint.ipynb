```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

import numpy as np
import random
import copy
from collections import deque, namedtuple

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
```

```python
# Models
def hidden_init(layer):
    fan_in = layer.weight.data.size()[0]
    lim = 1. / np.sqrt(fan_in)
    return (-lim, lim)

class Critic(nn.Module):
    def __init__(self, seed):
        super(Critic, self).__init__()
        self.seed = torch.manual_seed(seed)
        self.fc1 = nn.Linear(33, 400)
        self.fc2 = nn.Linear(404, 300)
        self.fc3 = nn.Linear(300, 1)
        self.reset_parameters()

    def reset_parameters(self):
        self.fc1.weight.data.uniform_(*hidden_init(self.fc1))
        self.fc2.weight.data.uniform_(*hidden_init(self.fc2))
        self.fc3.weight.data.uniform_(-3e-3, 3e-3)
        
    def forward(self, states, actions):
        x = F.relu(self.fc1(states))
        x = torch.cat((x, actions), -1)
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x
    
class Actor(nn.Module):
    def __init__(self, seed):
        super(Actor, self).__init__()
        self.seed = torch.manual_seed(seed)
        self.fc1 = nn.Linear(33, 400)
        self.fc2 = nn.Linear(400, 300)
        self.fc3 = nn.Linear(300, 4)
        self.reset_parameters()

    def reset_parameters(self):
        self.fc1.weight.data.uniform_(*hidden_init(self.fc1))
        self.fc2.weight.data.uniform_(*hidden_init(self.fc2))
        self.fc3.weight.data.uniform_(-3e-3, 3e-3)
        
    def forward(self, states):
        x = F.relu(self.fc1(states))
        x = F.relu(self.fc2(x))
        x = F.tanh(self.fc3(x))
        return x
```

```python
# Noise Function
class OUNoise:
    def __init__(self, size, mu=0., theta=0.15, sigma=0.2):
        self.mu = mu * np.ones(size)
        self.theta = theta
        self.sigma = sigma
        self.reset()

    def reset(self):
        self.state = copy.copy(self.mu)

    def sample(self):
        x = self.state
        dx = self.theta * (self.mu - x) + self.sigma * np.random.randn(len(x)) # ADDED
        self.state = x + dx
        return self.state
```

```python
# SumTree Function
class Node:
    def __init__(self, left, right, is_leaf: bool = False, idx = None):
        self.left = left
        self.right = right
        self.is_leaf = is_leaf
        if not self.is_leaf:
            self.value = self.left.value + self.right.value
        self.parent = None
        self.idx = idx
        if left is not None:
            left.parent = self
        if right is not None:
            right.parent = self
            
    @classmethod
    def create_leaf(cls, value, idx):
        leaf = cls(None, None, is_leaf=True, idx=idx)
        leaf.value = value
        return leaf
    
def create_tree(input: list):
    nodes = [Node.create_leaf(v, i) for i, v in enumerate(input)]
    leaf_nodes = nodes
    while len(nodes) > 1:
        inodes = iter(nodes)
        nodes = [Node(*pair) for pair in zip(inodes, inodes)]
    return nodes[0], leaf_nodes
    
def retrieve(value: float, node: Node):
    if node.is_leaf:
        return node.idx, node.value
    if node.left.value >= value:
        return retrieve(value, node.left)
    else:
        return retrieve(value - node.left.value, node.right)
        
def update(node: Node, new_value: float):
    change = new_value - node.value
    node.value = new_value[0]
    propagate_changes(change[0], node.parent)
        
def propagate_changes(change: float, node: Node):
    node.value += change
    if node.parent is not None:
        propagate_changes(change, node.parent)
```

```python
# Prioritized Replay Buffer
class PrioritizedReplayBuffer:
    def __init__(self, state_size, action_size, buffer_size, eps=1e-2, alpha=1, beta=1):
        self.eps = eps
        self.alpha = alpha
        self.beta = beta

        self.state = torch.empty(buffer_size, state_size, dtype=torch.float)
        self.action = torch.empty(buffer_size, action_size, dtype=torch.float)
        self.reward = torch.empty(buffer_size, dtype=torch.float)
        self.next_state = torch.empty(buffer_size, state_size, dtype=torch.float)
        self.done = torch.empty(buffer_size, dtype=torch.int)

        self.count = 0
        self.real_size = 0
        self.buffer_size = buffer_size
        self.root_node, self.leaf_nodes = create_tree(torch.ones(self.buffer_size))

    def add(self, transition):
        state, action, reward, next_state, done = transition
        self.state[self.count], self.action[self.count], self.reward[self.count], self.next_state[self.count], self.done[self.count] = \
        torch.as_tensor(state), torch.as_tensor(action), torch.as_tensor(reward), torch.as_tensor(next_state), torch.as_tensor(done)
        
        self.count = (self.count + 1) % self.buffer_size
        self.real_size = min(self.buffer_size, self.real_size + 1)
            
    def sample(self, batch_size):
        sample_idxs = []
        weights = []
        if self.real_size < 65536:
            sample_idxs = random.sample(range(self.real_size), batch_size)
            weights = torch.ones(batch_size) / batch_size
        else:
            tree_total = self.root_node.value
            for i in range(batch_size):
                rand_val = np.random.uniform(0, tree_total)
                sample_idx, mass = retrieve(rand_val, self.root_node)
                sample_idxs.append(sample_idx)
                prob = mass / tree_total
                weight = (self.real_size * prob) ** -self.beta
                weights.append(weight)

            weights = torch.tensor(weights)
            weights = weights/weights.max()

        batch = (
            self.state[sample_idxs],
            self.action[sample_idxs],
            self.reward[sample_idxs],
            self.next_state[sample_idxs],
            self.done[sample_idxs]
        )
        
        return batch, weights.unsqueeze(1), torch.tensor(sample_idxs).unsqueeze(1)

    def update_priorities(self, sample_idxs, td_error):
        priorities = td_error.detach()
        for sample_idx, priority in zip(sample_idxs, priorities):
            priority = (priority + self.eps) ** self.alpha
            update(self.leaf_nodes[sample_idx], priority)
```

```python
# Training Function
class Agent(nn.Module):
    def __init__(self, buffer_size, batch_size):
        super(Agent, self).__init__()
        self.buffer_size = buffer_size
        self.batch_size = batch_size
        self.dqn = Critic(2).to(device)
        self.dqn_target = Critic(2).to(device)
        self.policy = Actor(2).to(device)
        self.optimizer_dqn = optim.Adam(self.dqn.parameters(), lr = 1e-4)
        self.optimizer_policy = optim.Adam(self.policy.parameters(), lr = 1e-3)
        self.t_step = 0
        self.noise = OUNoise(4)
        self.memory = PrioritizedReplayBuffer(33, 4, self.buffer_size, 1e-2, 1, 0.5)
    
    def act(self, state, epsilon, add_noise=True):
        state = torch.from_numpy(state).float().to(device)
        self.policy.eval()
        with torch.no_grad():
            action = self.policy(state).cpu().data.numpy()
        self.policy.train()
        action += self.noise.sample() * epsilon
        return np.clip(action, -1, 1)
    
    def step(self, state, action, reward, next_state, done):
        e = [state, action, reward, next_state, done]
        self.memory.add(e) 
        
        self.t_step = (self.t_step + 1)
        if self.t_step % 20 <= 20:
            if self.memory.real_size>=self.batch_size:
                batch, weights, sample_idxs = self.memory.sample(self.batch_size)
                
                Q, td_error = self.dqn_learn(batch[0], batch[1], batch[2], batch[3], batch[4], weights)
                self.optimizer_dqn.zero_grad()
                Q.backward()
                self.optimizer_dqn.step()
                for target_param, local_param in zip(self.dqn_target.parameters(), self.dqn.parameters()):
                    target_param.data.copy_(0.001*local_param.data + (1.0-0.001)*target_param.data)
                self.memory.update_priorities(sample_idxs, td_error)
                
                P = self.policy_learn(batch[0], weights)
                self.optimizer_policy.zero_grad()
                P.backward()
                self.optimizer_policy.step()
                
    def dqn_learn(self, states, actions, rewards, next_states, dones, weights, discount = 0.99):
        next_actions = self.policy(next_states)
        dqn_future_reward = rewards.unsqueeze(1) + (discount * self.dqn_target(next_states, next_actions) * (1 - dones.unsqueeze(1)))
        dqn_expected_reward = self.dqn(states, actions)
        return torch.sum(weights * (dqn_future_reward - dqn_expected_reward)**2), abs(dqn_future_reward - dqn_expected_reward)
    
    def policy_learn(self, states, weights):
        actions_pred = self.policy(states)
        rewards_pred = self.dqn(states, actions_pred)
        return -torch.sum(weights * rewards_pred)
    
    def reset(self):
        self.noise.reset()

agent = Agent(65536, 128)
```

```python
# Training
def training(n_episodes=1000, max_t=1000, epsilon = 1.):
    scores = []
    scores_window = deque(maxlen=100)
    for i_episode in range(1, n_episodes+1):
        env_info = env.reset(train_mode = True)[brain_name]
        agent.reset()
        state = env_info.vector_observations[0]
        score = 0
        for t in range(max_t):
            if i_episode >= 150:
                epsilon *= 0.99
            action = agent.act(state, epsilon)
            env_info = env.step(action)[brain_name]           
            next_state, reward, done = env_info.vector_observations[0], env_info.rewards[0], env_info.local_done[0]
            score += env_info.rewards[0]                      
            agent.step(state, action, reward, next_state, done)
            state = next_state
            if done:
                break
        scores_window.append(score)
        scores.append(score)              
        if i_episode % 5 == 0:
            print('\rEpisode {}\tReward: {:.2f}\tAverage Reward: {:.2f}'.format(i_episode, score, np.mean(scores_window)))
        if i_episode % 100 == 0:
            print('\rEpisode {}\tAverage Score: {:.2f}'.format(i_episode, np.mean(scores_window)))
        if np.mean(scores_window)>=30.0:
            print('\nEnvironment solved in {:d} episodes!\tAverage Score: {:.2f}'.format(i_episode-100, np.mean(scores_window)))
            torch.save(agent.dqn.state_dict(), 'checkpoint/trained_weights_per_no_target_actor_critic.pth')
            torch.save(agent.policy.state_dict(), 'checkpoint/trained_weights_per_no_target_actor_policy.pth')
            break
    return scores

scores = training()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

# plot the scores
fig = plt.figure()
ax = fig.add_subplot(111)
plt.plot(np.arange(len(scores)), scores)
plt.ylabel('Score')
plt.xlabel('Episode #')
plt.show()
```
